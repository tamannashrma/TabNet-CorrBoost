```python
n=[]
tim = pd.DataFrame()
fname=[]
for file in os.listdir(loc_df_fs):
    
    df  = pd.read_csv(os.path.join(loc_df_fs,file))
    X  = df.iloc[:,1:-1].values   #change for df after fs only
    y  = df.iloc[:,-1].values
    
    fname.append(file.split('.')[0])
    
    
    t=[]
    
    for clf in lc:
        
        
        t1 = time.time()
        
        p = Pipeline([('smote',SVMSMOTE()),('norm',MinMaxScaler()),('clf',clf)])
        acc = cross_val_score(p,X,y,cv=cv,scoring ='accuracy')
        roc = cross_val_score(p,X,y,cv=cv,scoring ='roc_auc')
        f1 = cross_val_score(p,X,y,cv=cv,scoring ='average_precision')
        t2= time.time()
        
        t.append((t2-t1))
        
        
    tim = pd.concat([tim,pd.Series(t)],axis=1)
    
  
tim.index = ['SVM','NB','KNN','DT','RF','XGB','LGB','MLP']
tim.columns = fname    
    
#tim.to_csv(os.path.join(loc_rest,file))
    
```
