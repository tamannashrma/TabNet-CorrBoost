```python
import pandas as pd
import numpy as np
import os
from collections import Counter
```

```python
l1 = r"C:/Users/shubh/Desktop/1dcnn/AEEEM"
l2 = r"C:/Users/shubh/Desktop/1dcnn/AEEEM_boruta/"
l3 = r"C:\Users\shubh\Desktop\XGB DNN\df after fs\AEEEM"
```

```python
f1 =[]
f2=[]
f3 = []
fn =[]
r1=[]
r2 =[]
r3 =[]
o =[]
p=[]
for f in os.listdir(l1):
    df = pd.read_csv(os.path.join(l1,f))
    f1.append(df.shape[1])
    r1.append(df.shape[0])
    fn.append(f.split(".csv")[0])
    o.append(Counter(df.iloc[:,-1])[1])
    k=Counter(df.iloc[:,-1])[1]/df.shape[0]*100
    p.append(k)

for fff in os.listdir(l2):
    df = pd.read_csv(os.path.join(l2,fff))
    f3.append(df.shape[1]-1)
    r3.append(df.shape[0])
dframe = pd.concat([pd.Series(r1),pd.Series(o),pd.Series(p),pd.Series(f1),pd.Series(f3)],axis=1)
dframe.index =fn
dframe.columns =['No. of Samples','No. of Defective samples','Data Imbalance','Original No. of features','CorrBoost Features']
dframe.to_csv('ds analysis.csv')
```
