```python
loc_data = r"C:\Users\shubh\Desktop\GAN\Dataset\all"
res = "C:/Users/shubh/Desktop/GAN/Dataset/correlation/"
l=[]
m=[]
n=[]
for file in os.listdir(loc_data):
    if file.endswith(".csv"):
        data = pd.read_csv(os.path.join(loc_data,file))
        print(data.shape)
        l.append(data.shape[1])
        corr = data.corr()
        columns = np.full((corr.shape[0],), True, dtype=bool)
        for i in range(corr.shape[0]):
            for j in range(i+1, corr.shape[0]):
                if corr.iloc[i,j] >= 0.9:
                    if columns[j]:
                        columns[j] = False
        selected_columns = data.columns[columns]
        data1 = data[selected_columns]
        print(data1.shape)
        m.append(data1.shape[1])
        n.append(file.split(".csv")[0])
        data1.to_csv(os.path.join(res,file),index=False)
```

```python
dc = pd.concat([pd.Series(l),pd.Series(m)],axis=1)
dc.columns = ['Original features','Filtered Features']
dc.index = n
dc.to_csv("C:/Users/shubh/Desktop/GAN/Dataset/correlation/feature.csv")
```
