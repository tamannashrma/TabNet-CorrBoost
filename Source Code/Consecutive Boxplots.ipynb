```python
import pandas as pd
import numpy.random as rnd
import seaborn as sns
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8]
plt.rcParams['figure.dpi'] = 300
df = pd.read_csv(r"C:\Users\shubh\Desktop\XGB DNN\1407\analysis\boxplot\clf\ROC_clf.csv")
#df['Dataset'] = df.index.Dataset
df_plot = df.melt(id_vars='Dataset', value_vars=["Baseline", "FS"])
sns.set(font_scale = 1.5)
fig = sns.boxplot(x='Dataset', y='value', hue='variable', data=df_plot, color ='orange', fliersize =3,linewidth=2,
                 width=0.6,showmeans=True)
fig.set(xlabel ="Dataset", ylabel = "AUPRC", title ='Baseline vs FS Performance')
plt.savefig("AUPRC_ds.jpg")
```

```python
# Clasifier wise
import pandas as pd
import numpy.random as rnd
import seaborn as sns
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8]
plt.rcParams['figure.dpi'] = 300
df = pd.read_csv(r"C:\Users\shubh\Desktop\XGB DNN\Results\NASA\Analysis\NASA\boxplot\FS vs Baseline\clf wise\AUPRC_clf.csv")
#df['Dataset'] = df.index.Dataset
df_plot = df.melt(id_vars='Classifier', value_vars=["Baseline", "FS"])
sns.set(font_scale = 2)
fig = sns.boxplot(x='Classifier', y='value', hue='variable', data=df_plot, color ='red', fliersize =3,linewidth=2,
                 width=0.6,showmeans=True)
fig.set(xlabel ="Classifier", ylabel = "AUPRC", title ='Baseline vs FS Performance')
plt.xticks(rotation = 30)
plt.savefig(r"C:\Users\shubh\Desktop\CorrBoost\23122023\NASA_AUPRC_clf.jpg")
```
