```python
import os
import pandas as pd
res = "C:/Users/shubh/Desktop/XGB DNN/Results/AEEEM/xgb/"
res2 = "C:/Users/shubh/Desktop/XGB DNN/Results/AEEEM/Analysis/Baseline/"
m = "Accuracy" + "_baseline"
var1 = res2 + m + "_min" + ".csv"
var2 = res2 + m + "_mean"+ ".csv"
var3 = res2 + m + "_max" + ".csv"
```

```python
da=pd.DataFrame()
db=pd.DataFrame()
dc=pd.DataFrame()
file = []
for f in os.listdir(res):
    if f.endswith(".csv"):
        df = pd.read_csv((os.path.join(res,f)))
        dv1 = df.iloc[0:1,]
        dv2 = df.iloc[1:2,]
        dv3 = df.iloc[2:3,]
        da= pd.concat([da,dv1])
        db= pd.concat([db,dv2])
        dc= pd.concat([dc,dv3])
        file.append(f.split(".csv")[0])
da.index =file
db.index =file
dc.index =file
da=da.drop(da.columns[[0]],
                       axis = 1)
db=db.drop(db.columns[[0]],
                      axis = 1)
dc=dc.drop(dc.columns[[0]],
                       axis = 1)
da.to_csv(var1)
db.to_csv(var2)
dc.to_csv(var3)
        
```
